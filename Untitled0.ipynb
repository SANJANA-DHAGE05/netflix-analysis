In this step, we address missing data in the 'country' column by replacing all `NaN` values with the label 'Unknown'.
```python
df['country'].fillna('Unknown', inplace=True)
df.head(1)
```

```python
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
```

### Counting and Ranking the Most Common Netflix Genres

In this step, we extract the individual genres from the 'listed_in' column, which contains genre categories as comma-separated strings. By splitting these strings into lists and flattening them, we create a complete list of all genre entries. Using the `Counter` class, we count the frequency of each genre across the dataset. The resulting counts are then sorted in descending order to identify the top 10 most common genres on Netflix. This insight helps reveal which genres dominate the platform’s content library.


```python
from collections import Counter
all_genres = df['listed_in'].str.split(', ').sum()
genre_counts = pd.Series(Counter(all_genres)).sort_values(ascending=False)
print(genre_counts.head(10))
```

### Visualizing the Top 10 Netflix Genres

In this step, we present a bar chart of the ten most frequent genres found in the Netflix catalog. This visualization highlights which types of content are most prevalent on the platform, providing an immediate visual understanding of Netflix’s genre distribution and content strategy.


```python
#3. Visualize Top Genres

import matplotlib.pyplot as plt
genre_counts.head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Netflix Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.tight_layout()
plt.show()



```

### Analyzing Release Trends of Netflix Titles Over the Years

In this step, we examine how the number of Netflix titles released each year has changed over time. By grouping the data by 'release_year' and counting the number of titles per year, we can visualize trends in content production and acquisition. This line plot helps identify periods of significant growth or decline in Netflix’s library, revealing how the platform has evolved its content strategy, responded to global events, or expanded into new markets.

### Filtering Dataset for Movies Only

In this step, we create a subset of the dataset that includes only the rows where the 'type' is 'Movie'. This allows us to focus our analysis specifically on movie titles, which is useful for analyses like movie duration, genre distributions within movies, and other movie-specific insights.


```python
movies = df[df['type'] == 'Movie']
```

### Extracting Numeric Duration from Movie Duration Strings

The 'duration' column contains values like "90 min" or "1 Season" as strings. To analyze movie lengths quantitatively, we extract the numeric portion (minutes) using a regular expression. This step creates a new column, 'duration_num', converting the extracted digits to float type, which can then be used for statistical analysis and visualizations involving movie runtimes.


```python
# Extract digits (minutes) as float
movies['duration_num'] = movies['duration'].str.extract('(\d+)').astype(float)
```

### Calculating the Average Movie Duration

In this step, we compute the average duration of movies on Netflix using the numeric 'duration_num' column we previously created. This metric provides insight into typical movie length on the platform, which is useful for understanding content pacing and viewer engagement.


```python
avg_duration = movies['duration_num'].mean()
print(f'Average movie duration: {avg_duration:.1f} minutes')
```

### Visualizing the Distribution of Movie Durations

This step uses a histogram to display the distribution of movie durations on Netflix. By plotting the counts of movies across 20-minute bins, we can easily see the most common movie lengths, identify typical runtime ranges, and spot any outliers or unusual patterns. This visualization helps us understand Netflix’s typical content duration and assess whether the platform favors shorter, standard-length, or longer movies.


```python
import matplotlib.pyplot as plt

plt.hist(movies['duration_num'].dropna(), bins=20, color='coral', edgecolor='black')
plt.title('Distribution of Movie Durations')
plt.xlabel('Minutes')
plt.ylabel('Number of Movies')
plt.show()
```

### Identifying the Top Content-Producing Countries

In this step, we analyze where Netflix content originates by examining the 'country' column. We first replace any missing values with an empty string to avoid processing errors. Next, we split the comma-separated country strings for each title into lists, flatten all these lists into a single collection, and count the frequency of each country’s appearance in the dataset using Python’s `Counter` class. Sorting these counts in descending order allows us to identify the top 10 countries that contribute the most content to the Netflix catalog. This insight reveals Netflix’s strongest production and licensing markets.



```python
from collections import Counter
import pandas as pd

# Fill NaNs with empty string first!
df['country'] = df['country'].fillna('')

# Split and flatten all country lists correctly
all_countries = df['country'].str.split(', ').sum()

# Now count
country_counts = pd.Series(Counter(all_countries)).sort_values(ascending=False)
print(country_counts.head(10))
```

### Creating a Primary Genre Column and Visualizing Top Genres

In this step, we create a new 'genre' column by extracting the first genre listed for each title in the 'listed_in' column. This simplifies subsequent genre analysis by focusing on the primary genre associated with each show or movie.

We then visualize the top 5 most common genres on Netflix by counting how often each primary genre appears and displaying the results in a bar chart. This plot highlights which genre categories dominate the platform’s catalog and provides quick insight into Netflix’s content strategy.


```python
# Create a new 'genre' column with the first listed genre
df['genre'] = df['listed_in'].str.split(',').str[0]

top_genres = df['listed_in'].value_counts().head(5)
top_genres.plot(kind='bar', color='teal')
plt.title('Top 5 Netflix Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

### Visualizing the Distribution of Genres with a Pie Chart

In this step, we use a pie chart to display the proportion of each primary genre in the Netflix catalog. By calculating the value counts of the 'genre' column, we obtain the frequency of each genre’s appearance as the main category for titles. The pie chart visually represents how content is distributed across genres, making it easy to grasp which genres are most and least prevalent on the platform.


```python
genre_dist = df['genre'].value_counts()
plt.pie(genre_dist, labels=genre_dist.index, autopct='%1.1f%%', startangle=140)
plt.title('Genre Distribution')
plt.axis('equal')
plt.show()
```

### Visualizing Correlations Between Numerical Features with a Heatmap

In this step, we use a correlation heatmap to explore relationships between all numerical columns in the dataset. We first select only the numeric columns and compute their pairwise correlation coefficients. The seaborn `heatmap` visually displays these correlations, where darker shades indicate stronger positive or negative relationships between features. Annotating the heatmap allows us to quickly spot variables that are highly correlated, which is useful for deeper statistical analysis and for identifying patterns or potential issues such as multicollinearity in predictive modeling.


```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap of Numerical Features')
plt.show()


```

### Visualizing the Trend of Netflix Releases Over the Years with an Area Chart

In this step, we use an area chart to visualize how the number of Netflix titles released per year has evolved over time.
